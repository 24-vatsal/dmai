```python
def calculate_support(candidate_itemsets, dataset_df):
    """Calculates the support of each candidate itemset in the dataset."""
    itemset_support = {}
    num_transactions = len(dataset_df)

    for candidate_itemset in candidate_itemsets:
        count = 0
        for index, row in dataset_df.iterrows():
            is_present = True
            for item in candidate_itemset:
                if not row[item]:
                    is_present = False
                    break
            if is_present:
                count += 1
        itemset_support[candidate_itemset] = count / num_transactions
    return itemset_support
```

```python
def find_frequent_itemsets(df, min_support):
    """Finds frequent itemsets based on the minimum support threshold."""
    frequent_itemsets = {}
    k = 1

    # Find frequent 1-itemsets
    frequent_itemsets[k] = []
    for col in df.columns:
        support = df[col].sum() / len(df)
        if support >= min_support:
            frequent_itemsets[k].append((col,))

    # Iteratively find frequent k-itemsets
    while frequent_itemsets[k]:
        k += 1
        candidate_itemsets_k = generate_candidates(frequent_itemsets[k-1], k)
        support_k = calculate_support(candidate_itemsets_k, df)
        frequent_itemsets[k] = [itemset for itemset, support in support_k.items() if support >= min_support]

    return frequent_itemsets
```

```python
from itertools import combinations

def generate_association_rules(frequent_itemsets, min_confidence):
    """Generates association rules from frequent itemsets based on minimum confidence."""
    association_rules = []


    all_itemset_supports = {}


    def get_support(itemset, df):
        """Calculates the support of a given itemset in the dataset."""
        count = 0
        num_transactions = len(df)
        if not itemset:
             return 1.0
        for index, row in df.iterrows():
            is_present = True
            for item in itemset:

                if item not in df.columns or not row[item]:
                    is_present = False
                    break
            if is_present:
                count += 1
        return count / num_transactions



    for k in sorted(frequent_itemsets.keys()):
        if k < 2:
            continue

        for frequent_itemset_tuple in frequent_itemsets[k]:
            frequent_itemset = set(frequent_itemset_tuple)
            frequent_itemset_support = get_support(frequent_itemset, df)


            for i in range(1, k):
                for antecedent_tuple in combinations(frequent_itemset_tuple, i):
                    antecedent = set(antecedent_tuple)
                    consequent = frequent_itemset - antecedent


                    if not consequent:
                        continue

                    antecedent_support = get_support(antecedent, df)


                    if antecedent_support == 0:
                        continue

                    confidence = frequent_itemset_support / antecedent_support

                    if confidence >= min_confidence:
                        association_rules.append((tuple(sorted(list(antecedent))), tuple(sorted(list(consequent))), confidence))

    return association_rules
```
